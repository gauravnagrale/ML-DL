# Creating the RFM metrics and RFM score

```python
df_clean.tail()
```

```python
today_date = dt.datetime(2011, 12, 9)
today_date
```

```python
df_clean['Date'] = pd.to_datetime(df_clean['Date'])
```

```python
rfm = df_clean.groupby("CustomerID").agg({"Date": lambda x: (today_date - x.max()).days,
                                "ï»¿InvoiceNo": lambda x: len(x),
                               "Total Revenue": lambda x: x.sum()})
rfm.columns = ["recency", "frequency", "monetary"]


# .days is used here to calculate the days between the two given dates
#EX:
'''d0 = date(2011, 1, 8)
d1 = date(2015, 8, 16)
delta = d1 - d0
print('The number of days between the given range of dates is :')
print(delta.days)'''



#-ve value is not telling us anything
rfm = rfm[rfm["monetary"] > 0]
rfm.head()
```

```python
#as we can see that the data is highely skewed we have to normalize the data before we further calculate the RFM score
rfm['R_rank'] = rfm['recency'].rank(ascending=False)
rfm['F_rank'] = rfm['frequency'].rank(ascending=True)
rfm['M_rank'] = rfm['monetary'].rank(ascending=True)
 
# normalizing the rank of the customers
rfm['R_norm'] = (rfm['R_rank']/rfm['R_rank'].max())*100
rfm['F_norm'] = (rfm['F_rank']/rfm['F_rank'].max())*100
rfm['M_norm'] = (rfm['F_rank']/rfm['M_rank'].max())*100
 
#rfm.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)
#rfm.head()
print("*"*50)
print("displaying the R-rank data",rfm['R_rank'].head())
print("displaying the F-rank data",rfm['F_rank'].head())
print("displaying the M-rank data",rfm['M_rank'].head())
print("*"*50)
print("displaying the Normalize data data",rfm['R_norm'].head())
print("displaying the Normalize data data",rfm['F_norm'].head())
print("displaying the Normalize data data",rfm['M_norm'].head())
print("*"*50)
print("displaying the data",rfm.head())
```

# calulating the RFM score

```python
#as recency is the inversely proportional ie as the recency decreases the rfm score increases 
rfm['RFM_score']=(rfm['R_norm'])*0.15+(rfm['F_norm']*0.28)+(rfm['M_norm']*0.57)
rfm['RFM_score'] *= 0.05
rfm = rfm.round(2)
rfm
```

```python
rfm['RFM_score'].describe()
```

# customer segmentation 

Champions - bought recently, buy often and spend the most

Loyal Customers - spend good money and often, responsive to promotions

Potential Loyalist - recent customers, but spent a good amount and bought more than once

New Customers - bought most recently, but not often

Promising - recent shoppers, but haven’t spent much

Needing Attention - above average recency, frequency and monetary values; may not have bought very recently though

About To Sleep - below average recency, frequency and monetary values; will lose them if not reactivated

At Risk - spent big money and purchased often but long time ago; need to bring them back

Can't Loose Them - made biggest purchases, and often but haven’t returned for a long time

Hibernating - last purchase was long back, low spenders and low number of orders

# example of the Where 

```python
#1Create a Numpy array from a list
arr = np.array([11, 12, 13, 14])
high_values = ['High', 'High', 'High', 'High']
low_values = ['Low', 'Low', 'Low', 'Low']
# numpy where() with condition argument
result = np.where(arr > 12,['High', 'High', 'High', 'High'],['Low', 'Low', 'Low', 'Low'])
#here after condition gets true then the first list value will be taken and if the condition fails the list 2 values will be taken
print("First result:",result)

#2
arr = np.array([11, 12, 14, 15, 16, 17])
# pass condition expression only
result = np.where((arr > 12) & (arr < 16),['A', 'A', 'A', 'A', 'A', 'A'],['B', 'B', 'B', 'B', 'B', 'B'])
print("Second result:",result)
```

```python
rfm["Customer_segment"] = np.where(rfm['RFM_score'] >4.5,"Champions", 
                                   (np.where(rfm['RFM_score'] > 4,"potential_loyalists",
                                        (np.where(rfm['RFM_score'] > 3,"promising",
                                         (np.where(rfm['RFM_score'] > 2,"need_attention",
                                        (np.where(rfm['RFM_score'] > 1,'cant_loose',
                                        np.where(rfm['RFM_score'] > 0,'hibernating', 'Lost Customers'))))))))))
rfm.head()
```

# Modeling:

# 1)K-means algo

```python
rfm_prep['RFM_score']=round(rfm_prep['RFM_score'])
rfm_prep.head()
```

```python
x=rfm_prep.iloc[:,0:4].values
x
```

```python
#importing the KMeans model
from sklearn.cluster import KMeans
```

```python
#Using the Elbow method to find the optimal value of k
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
#PLotting the Elbow-Curve
plt.plot(range(1,11),wcss)
plt.title('The Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

# we can see from the Elbow curve that no of cluster we want to form is 6

```python
#re-building the model with the no of clusters=6
kmeans=KMeans(n_clusters=6,init="k-means++",random_state=42)
```

```python
#predicting the value
y_mean=kmeans.fit_predict(x)
y_mean
```

# visualizing the cluster

```python
plt.figure(figsize=(7,5))
plt.scatter(x[y_mean == 0,0],x[y_mean == 0,1],s=100,c='red',label='cluster1')
plt.scatter(x[y_mean == 1,0],x[y_mean == 1,1],s=100,c='blue',label='cluster2')
plt.scatter(x[y_mean == 2,0],x[y_mean == 2,1],s=100,c='green',label='cluster3')
plt.scatter(x[y_mean == 3,0],x[y_mean == 3,1],s=100,c='cyan',label='cluster4')
plt.scatter(x[y_mean == 4,0],x[y_mean == 4,1],s=100,c='black',label='cluster5')
plt.scatter(x[y_mean == 5,0],x[y_mean == 5,1],s=100,c='magenta',label='cluster6')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroyed')
plt.title('Cluster of Customer')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.show()
```
